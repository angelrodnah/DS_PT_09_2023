```python
import hashlib
import requests
import datetime

timestamp = datetime.datetime.now().strftime('%Y-%m-%d%H:%M:%S')
pub_key = '758a6ce65a83bd9d58da100f5e98bc87'
priv_key = 'f63a5e5176e048a39d38e629658bf1d4ca081b77'

def hash_params():
    hash_md5 = hashlib.md5()
    hash_md5.update(f'{timestamp}{priv_key}{pub_key}'.encode('utf-8'))
    hashed_params = hash_md5.hexdigest()
    return hashed_params

params = {'limit':100,
          'nameStartsWith':'A', 
          'ts': timestamp, 
          'apikey': pub_key, 
          'hash': hash_params()}

url='http://gateway.marvel.com/v1/public/characters'

res = requests.get(url,params=params)

resultado = res.json()
```

```python
data=resultado['data'][ 'results']
```

```python
heroes_df=pd.DataFrame(data)
heroes_df.columns
```

```python
heroes=pd.DataFrame()
heroes['Id']=heroes_df.id
heroes['Nombre']=heroes_df.name
heroes['Imagen']=heroes_df['thumbnail'].apply(lambda d:"".join([d['path'],'.',d['extension']]))
heroes['Comics']=heroes_df['comics'].apply(lambda d:"; ".join([d.get('name') for d in d.get('items')]))
```

```python
import pandas as pd
from IPython.display import HTML

def Display(df, RutaImagen='Imagen'):
    def path_to_image_html(path):
        return '<img src="' + path + '" width="200" />'
    return HTML(df.to_html(escape=False, formatters={RutaImagen: path_to_image_html}))


```
